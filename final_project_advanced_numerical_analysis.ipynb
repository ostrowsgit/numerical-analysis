# 3.
Acoustic waves. The air pressure $p(x, t)$ in an organ pipe is governed by the wave equation 
$$\frac{\partial^2p}{\partial t^2}=c^2\frac{\partial^2p}{\partial x^2}\quad\quad 0<x<l,\quad\quad t>0$$

where $l$ is the length of the pipe and $c$ is a constant. If the pipe is open, the boundary conditions are given by
$$p(0,t)=p_0\quad\quad \textrm{ and } p(l,t)=p_0.$$
If the pupe is closed at the end $x=l$ the boundary conditions are
$$p(0,t)=p_0\quad\quad \textrm{ and } \frac{\partial p}{\partial x}p(l,t)=0.$$
Assume that $c=1, l=1$, and the initial conditions are
$$p(x,0)=p_0\cos(2\pi x)\quad\quad \textrm{ and } \frac{\partial p}{\partial x}p(x,0)=0\quad\quad 0\leq x\leq1.$$

## 3.a.

Write down an explicit finite difference method for $(5)$ and give stability conditions and the order of the method.

Consider the finite difference method to approximate $P_{tt}=c^2P_{xx}$

$$\frac{p_j^{n+1}-2p_j^n+p_j^{n-1}}{k^2} = c^2 \frac{p_{j-1}^n-2p_j^n+p_{j+1}^n}{h^2}\quad\quad (*)$$

with $c=1$ and $\lambda = \frac kh$ the explicit form becomes
$$p_j^{n+1}=2p_j^n-p_j^{n-1}+\frac{k^2}{h^2}(p_{j-1}^n-2p_j^n+p_{j+1}^n)\\
p_j^{n+1}=\lambda^2 p_{j-1}^n+2(p_j^n-\lambda^2 p_j^n)+\lambda^2 p_{j+1}^n-p_j^{n-1}$$

we find the order of the finite difference method by finding the truncation error $\;t_j^n(k,h)$ for $\Delta t = k, \Delta x = h$.

$$\begin{align}
P(t_n,x_{j+1})= P(t_n,x_j+h) &= P(t_n,x_j)+P_x(t_n,x_j)h+P_{xx}(t_n,x_j)\frac{h^2}{2!}+P_{xxx}(t_n,x_j)\frac{h^3}{3!}+\dots&\quad\quad(1)\\
P(t_n,x_{j-1}) = P(t_n,x_j-h) &= P(t_n,x_j)-P_x(t_n,x_j)h+P_{xx}(t_n,x_j)\frac{h^2}{2!}-P_{xxx}(t_n,x_j)\frac{h^3}{3!}+\dots&\quad\quad(2)\\
P(t_{n+1},x_j) = P(t_n+k,x_j) &= P(t_n,x_j)+P_t(t_n,x_j)k+P_{tt}(t_n,x_j)\frac{k^2}{2!}+P_{ttt}(t_n,x_j)\frac{k^3}{3!}+\dots&\quad\quad(3)\\
P(t_{n-1},x_j)=P(t_n-k,x_j) &= P(t_n,x_j)-P_{t}(t_n,x_j)k+P_{tt}(t_n,x_j)\frac{k^2}{2!}-P_{ttt}(t_n,x_j)\frac{k^3}{3!}+\dots&\quad\quad(4)\\
\textrm{then the second central }&\textrm{difference formulas with respect to } h \textrm{ and } k \textrm{ are}& \\
D^2_{h}P(t_n,x_j)=\frac{1}{h^2}((1)-2P(t_n,x_j)+(2)) &= 2P_{xx}(t_n,x_j)\frac{1}{2}+2P_{xxxx}(t_n,x_j)\frac{h^2}{4!}+\dots&\quad\quad(5)\\
D^2_{k}P(t_n,x_j)=\frac{1}{k^2}((1)-2P(t_n,x_j)+(2)) &= 2P_{tt}(t_n,x_j)\frac{1}{2}+2P_{tttt}(t_n,x_j)\frac{k^2}{4!}+\dots&\quad\quad(6)\\
\textrm{then truncation error } t_j^{n}(k,h)&=(6)-(5)\\
&=c_4^*(k^2+h^2)+\mathcal O(k^4)+\mathcal O(h^4) \textrm{ for some real } c_4^*\\
&= \mathcal O(k^2)+\mathcal O(h^2)\\
\end{align}$$
Then the finite difference method $(*)$ is of order $\mathcal O(\Delta t)^2+\mathcal O(\Delta x)^2$.

To find the stability conditions of the method we apply the Von Neumann Analysis
with $P_j^n=g^ne^{ijmh}$ to $(*)$

$$\begin{align}
\frac{1}{k^2}(g^{n+1}e^{ijmh}-2g^ne^{ijmh}+g^{n-1}e^{ijmh})&=\frac{1}{h^2}(g^ne^{i(j-1)mh}-2g^ne^{ijmh}+g^ne^{i(j+1)mh})\\
\textrm{with }\lambda=\frac kh \textrm{ and dividing by for } g^ne^{ijmh}\\
g-2+\frac 1g &=\lambda (e^{-imh}-2+e^{imh})\\
\textrm{ with } e^{imh}+e^{-imh} &= 2\cos(mh)\\
 \textrm{ then with the half angle formula } 2\cos(mh)-2 & = -2(1-\cos(mh))=-2\left(2\left(\sin^2\left({\frac{mh}{2}}\right)\right)\right) \\
\textrm{then }g-2+\frac 1g &=-4\sin^2\left( \frac{mh}{2}\right)\\
\textrm{becomes }\left(\sqrt{g}-\frac {1}{\sqrt{g}}\right)^2 &=\left(\pm2i\lambda \sin\left( \frac{mh}{2}\right)\right)^2\\
\sqrt{g}-\frac {1}{\sqrt{g}} &=\pm2i\lambda \sin\left( \frac{mh}{2}\right) \textrm{ then multiplying by }\sqrt{g} \textrm{ we get}\\
g \pm 2i\lambda\sin\left(\frac{mh}{2}\right)\sqrt{g}-1&=0 \textrm{ which is quadratic in } \sqrt{g}\\
\textrm{then }\sqrt{g}_{\pm} &= \pm i\lambda\sin\left(\frac{mh}{2}\right)\pm\sqrt{1-\lambda^2\sin\left(\frac{mh}{2}\right)}\\
\textrm{then }g_{\pm} &= \left(\pm i\lambda\sin\left(\frac{mh}{2}\right)\pm\sqrt{1-\lambda^2\sin\left(\frac{mh}{2}\right)}\right)^2\\
&=-\lambda^2\sin^2\left(\frac{mh}{2}\right) \pm 2i\lambda\sin\left(\frac{mh}{2}\right)\sqrt{1-\lambda^2\sin\left(\frac{mh}{2}\right)} \\
&+\left(1-\lambda^2\sin\left(\frac{mh}{2}\right)\right)\\
\textrm{then taking the modulus of }&g \textrm{ we get }\\
\left|g_{\pm}\right|&=\lambda^4\sin^2\left(\frac{mh}{2}\right) +1-2\lambda^2\sin^2\left(\frac{mh}{2}\right)+\lambda^4\sin^2\left(\frac{mh}{2}\right)\\
&-2\lambda^2\sin^2\left(\frac{mh}{2}\right)\left( 1-\lambda^2\sin\left(\frac{mh}{2}\right) \right)\\
&+1-2\lambda^2\sin\left(\frac{mh}{2}\right)+\lambda^4\sin^2\left(\frac{mh}{2}\right)\\
&-4\lambda^2\sin^2\left(\frac{mh}{2}\right)\left( 1-\lambda^2\sin^2\left(\frac{mh}{2}\right) \right)\\
&\leq \lambda^4 +1 -2\lambda^2+\lambda^4-2\lambda^2(1-\lambda^2)\\
&+1-2\lambda^2 +\lambda^4-\lambda^4(1-\lambda^2)\\
&=\lambda^6+4\lambda^4-6\lambda^2+2\\
\textrm{then } |g_{\pm}| &\leq 1 \textrm{ if and only if } |\lambda|\leq 1\\
\textrm{for the } g_+=g_- \textrm{ solution, if  }mh&=0,\pi \textrm{ and }|\lambda|=1 \\
\textrm{then } u_j^n \textrm{ has linear growth  } n &\rightarrow\ \infty \textrm{ which grows slower than }g^{n-1} \\
 \textrm{ since the finite difference method}  &\textrm{ is of second order in }t 
\end{align}$$

Thus, the the method is stable if and only if $|\lambda| \leq 1$ 

which is the CFL condition $|a|\frac{\Delta t}{\Delta x}\leq 1$ with $a=c^2$.

We find initial condition for $n=1$. With $\frac{\partial p}{\partial t}(x,0)=0$ from $(9)$ we approximate $p(1,x) \approx P_1(x)$. We use the forward finite difference formula  $P_1(x)=\frac{p_j^1(x_j)-p_j^0(x_j)}{\Delta t}=0$ for $j = 1,\dots, M-1$ with respect to $t$.  Then $p_j^1(x_j)=p_j^0(x_j)$ for $j = 1,\dots, M-1$

```python
'''
WAVE EQUATION OF SECOND ORDER EQUATIONS FINITE DIFFERENCE METHOD waveEq2ndOrdersFD():
INPUT:,pj,pjPlus1,pjnMinus1,lmbda
'pj' is the previous space step's and current x step's approximation node,
'pjMinus1' is the previous time step's and next x step's approximation node,
'pjPlus1' is the next tim
'lmbda' is Courant number
OUTPUT: finite difference scheme forward in time and space

INITIAL VALUE FUNCTION i): u_0ANDu_1()
INPUT:
'x' a value from the domain of function for approximation,
OUTPUT: initial values at initial time step t = 0


GHOST POINT APPROXIMATION OF RIGHT BOUNDARY POINT ghostPointFD():
INPUT:
'pMminus1','pM','lmbda' defined above,
'pMnMinus1' 2 time steps agoth current x step's approximation node,
OUTPUT: an approximation of the right boundary point 

OPEN PIPE AIR PRESSURE APPROXIMATION openPipeAirPressure():
'p0' is the initial value of the boundary,
'h' derivative of approximation with respect to space variable 'x',
'k' derivative of approximation with respect to time variable 't',
'l' is the length of space of the pipe,
'T' is the length of time

CONSTANTS, ARRAYS AND VARIABLES:
'N' number of time steps,
'M' number of mesh grid steps,
'delt' cumulative time per iteration
'''

import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
import mpl_toolkits 
from matplotlib import rc
from IPython.display import display, Math, Latex

# Say, "the default sans-serif font is Palatino Linotype"
matplotlib.rcParams['font.serif'] = "Palatino Linotype"
# Then, "ALWAYS use serif fonts"
matplotlib.rcParams['font.family'] = "DejaVu Sans"
# set the default Latex font for Latex text
plt.rcParams["mathtext.fontset"] = "cm"
# set default font size
matplotlib.rcParams.update({'font.size': 16})
```

```python
def waveEq2ndOrdersFD(pjMinus1,pj,pjPlus1,pjnMinus1,lmbda):
  return( lmbda**2*pjMinus1+2*(1-lmbda**2)*pj+lmbda**2*pjPlus1-pjnMinus1 )

def u_0ANDu_1(x,p0):
  return (p0*np.cos(2*np.pi*x))

p0=0.9
k=0.05;h=k;l=1.0;lmbda=k/h
N=20;M=20;T=k*N

x = np.linspace(0.0,l,M+1) # domain of x

P = np.full([N+1, M+1], p0) # initialize with boundary conditions

# fill first inner nodes initial value with u_0 till the boundary condition
P[0,1:-1] = [u_0ANDu_1(x[j],p0) for j in range(1,M)]; P[1,:] = P[0,:]

for n in range(1,N):
  P[n+1,1:-1] = [waveEq2ndOrdersFD(P[n,j-1],P[n,j],P[n,j+1],P[n-1,j],lmbda) for j in range(1,M)]
```

```python
for t in range(10,N+1,10):
    delt=k*t
    fig = plt.figure()
    ax = fig.add_subplot (111)
    ax.plot(x,P[t,:])
    ax.set_title("$M=%i,\\; T=%f,\\; N=%i, \\; \\lambda = %f,t = %f$" %(M,T,N,lmbda,delt))
    plt.ylabel('$p(x,t)$')
    plt.xlabel('$x$')
plt.show()
```

## 3.c.
Implement your method given in $a)$ for the closed pipe at $x = l$ with $p_0 = 0.9$, and with step sizes $k = h = 0.05$. Plot your numerical solution at $t = 0.5$ and $t = 1.0$.


When the pipe is closed with $x=l\;$ we find the approximation for the ghost point $p_{M+1}^n$ in

$$\frac{p_M^{n+1}-2p_M^n+p_M^{n-1}}{(\Delta t)^2}=\frac{p_{M+1}^n-2p_m^n+p_{M-1}^n}{(\Delta x)^2}\quad\quad\quad (**)$$
with ghost point $x_{M+1}=x_M+\Delta x = 1+h$ outside the domain.

With closed pipe boundary condition for $x=l$ we use Centered Difference formula by $(7.10)$ 
$$\frac{\partial p}{\partial x}(x_M,t_n) = 0 \approx \frac{p_{M+1}^n-p_{M-1}^n}{2\Delta x}$$
then $p_{M+1}^n=p_{M-1}^n$. 

This implies that $(**)$ becomes
$$\frac{p_M^{n+1}-2p_M^n+p_M^{n-1}}{(\Delta t)^2}=2\frac{p_{M-1}^n-p_M^n}{(\Delta x)^2}$$ 
then the explicit form is

$$p_M^{n+1}=2\lambda^2p_{M-1}^n+2(1-\lambda^2)p_M^n-p_M^{n-1}$$

```python
def ghostPointFD(pMminus1,pM,pMnMinus1,lmbda):
  return( 2*lmbda**2*pMminus1+2*(1-lmbda**2)*pM-pMnMinus1 )

P = np.full([N+1, M+1], p0) # initialize with left boundary condition

# fill first inner nodes initial value with u_0 till the boundary condition
P[0,1:-1] = [u_0ANDu_1(x[j],p0) for j in range(1,M)]; P[1,:] = P[0,:]

for n in range(1,N):
  P[n+1,1:-1] = [waveEq2ndOrdersFD(P[n,j-1],P[n,j],P[n,j+1],
                                   P[n-1,j],lmbda) for j in range(1,M)]
  P[n+1,M] = ghostPointFD(P[n,M-1],P[n,M],P[n-1,M],lmbda)
```

```python
for t in range(10,N+1,10):
    delt=k*t
    fig = plt.figure()
    ax = fig.add_subplot (111)
    ax.plot(x,P[t,:])
    ax.set_title("$M=%i,\\; T=%f,\\; N=%i, \\; \\lambda = %f,t = %f$" %(M,T,N,lmbda,delt))
    plt.ylabel('$p(x,t)$')
    plt.xlabel('$x$')
plt.show()
```

## 3.d.

Repeat $b)$ for $k = h = 0.025$. Construct a higher order approximation by extrapolating your numerical solutions corresponding to $k = h = 0.05$ and $k = h = 0.025$. 

We construct an approximation by Richardson Extrapolation $(7.3)$ in the lecture notes

$$ p^{ext}(t_n,x_j) = \frac13 \big(4p(t_{2n},x_{2j}) -p(t_n,x_j)\big)$$ for $n=0,\dots. N$ and $j = 0,\dots,M $.

```python
def waveEq2ndOrdersFD(pjMinus1,pj,pjPlus1,pjnMinus1,lmbda):
  return( lmbda**2*pjMinus1+2*(1-lmbda**2)*pj+lmbda**2*pjPlus1-pjnMinus1 )

def u_0ANDu_1(x,p0):
  return (p0*np.cos(2*np.pi*x))

p0=0.9
k1=0.05;h1=k1;l=1.0;T=1.0
Pext = np.full([N+1, M+1], p0) 

def openPipeAirPressure(p0,h,k,l,T):
  N=int(T/k);M=int(l/h);lmbda=k/h
  
  x = np.linspace(0.0,l,M+1) # domain of x

  P = np.full([N+1, M+1], p0) # initialize with boundary conditions
    
  # fill first inner nodes initial value with u_0 till the boundary condition
  P[0,1:-1] = [u_0ANDu_1(x[j],p0) for j in range(1,M)]; P[1,:] = P[0,:]

  for n in range(1,N):
    P[n+1,1:-1] = [waveEq2ndOrdersFD(P[n,j-1],P[n,j],
                                     P[n,j+1],P[n-1,j],lmbda) for j in range(1,M)]

  return(P, x)

Pn,   xM = openPipeAirPressure(p0,h1, k1,l,T)
P2n, x2M = openPipeAirPressure(p0,h1/2,k1/2,l,T)
P2n = P2n[::2,::2]

Pext[:,:] = 1/3.0*(4.0*P2n[:,:]-Pn[:,:])
```

```python
for t in range(10,N+1,10):
    delt=k*t
    fig = plt.figure()
    ax = fig.add_subplot (111)
    ax.plot(x,Pext[t,:])
    ax.set_title("$M=%i,\\; T=%f,\\; N=%i, \\; \\lambda = %f,t = %f$" %(M,T,N,lmbda,delt))
    plt.ylabel('$p(x,t)$')
    plt.xlabel('$x$')
plt.show()
```

What's the order of the new approximation?

Let $\tilde P(t_{2n},x_{2j})$ for $j = 1,\dots, M$ and $n = 1,\dots, N$ be the approximation with double the space and time

points of the original approximation $P(t_n,x_j)$ with only the evenly indexed terms. Then asymptotically
$$\begin{align}
\tilde P(t_{2n},x_{2(j+1)})=P\left(t_n,x_j+\frac h2\right) &= P(t_n,x_j)+c_1P_x(t_n,x_j)\frac h2+c_2P_{xx}(t_n,x_j)\frac{\left(\frac h2\right)^2}{2!}+c_3P_{xxx}(t_n,x_j)\frac{\left(\frac h2\right)^3}{3!}+\dots&\quad\quad(7)\\
\tilde P(t_{2n},x_{2(j-1)})=P\left(t_n,x_j-\frac h2\right) &= P(t_n,x_j)-c_1P_x(t_n,x_j)\frac h2+c_2P_{xx}(t_n,x_j)\frac{\left(\frac h2\right)^2}{2!}-c_3P_{xxx}(t_n,x_j)\frac{\left(\frac h2\right)^3}{3!}+\dots&\quad\quad(8)\\
\tilde P(t_{2(n+1)},x_{2j})=P\left(t_n+\frac k2,x_j\right) &= P(t_n,x_j)+c_1P_t(t_n,x_j)\frac k2+c_2P_{tt}(t_n,x_j)\frac{\left(\frac k2\right)^2}{2!}+c_3P_{ttt}(t_n,x_j)\frac{\left(\frac k2\right)^3}{3!}+\dots&\quad\quad(9)\\
\tilde P(t_{2(n-1)},x_{2j})=P\left(t_n-\frac k2,x_j\right) &= P(t_n,x_j)-c_1P_{t}(t_n,x_j)\frac k2+c_2P_{tt}(t_n,x_j)\frac{\left(\frac k2\right)^2}{2!}-c_3P_{ttt}(t_n,x_j)\frac{\left(\frac k2\right)^3}{3!}+\dots&\quad\quad(10)\\
\textrm{then the second central }&\textrm{difference formulas with respect to } \frac h2 \textrm{ and } \frac k2 \textrm{ are}& \\
D^2_{h/2}P(t_n,x_j)=\frac{1}{\left(\frac h2\right)^2}((7)-2P(t_n,x_j)+(8)) &= 2c_2P_{xx}(t_n,x_j)\frac{1}{2}+2c_4P_{xxxx}(t_n,x_j)\frac{\left(\frac h2\right)^2}{4!}+\dots&\quad\quad(11)\\
D^2_{k/2}P(t_n,x_j)=\frac{1}{\left(\frac k2\right)^2}((9)-2P(t_n,x_j)+(10)) &= 2c_2P_{tt}(t_n,x_j)\frac{1}{2}+2c_4P_{tttt}(t_n,x_j)\frac{\left(\frac k2\right)^2}{4!}+\dots&\quad\quad(12)\\
\textrm{ with }c_i \textrm{ being real coefficients} \textrm{ for } i= 2,4,6,\dots\\
\textrm{then the extrapolation term with respect to }h\\
\textrm{ is } D^2_hP(t_n,x_j) &= \frac 13 (4D^2_{h/2}P(t_n,x_j)-D^2_{h}P(t_n,x_j)) = P_{xx}(t_n,x_j)+\tilde c_4h^4 +\mathcal O(h^6)\\
\textrm{ and with respect to } k \textrm{ is }  D^2_kP(t_n,x_j) &= \frac 13 (4D^2_{k/2}P(t_n,x_j)-D^2_{k}P(t_n,x_j)) = P_{tt}(t_n,x_j)+\ddot{\breve c}_4k^4 +\mathcal O(k^6)\\
\textrm{then the truncation error of the }&\textrm{Richardson Extrapolation is }\\
\tau^{ext}(t_n,x_j) = D^2_kP(t_n,x_j)-D^2_hP(t_n,x_j) &= P_{tt}(t_n,x_j)-P_{xx}(t_n,x_j)+\tilde{c}_4(k^4+h^4)+\mathcal O(k^6+h^6) \\
&=\tilde{\ddot c}_4(k^4+h^4)+\mathcal O(k^6+h^6)\\
\textrm{ for real coefficients } \tilde{c}_4, \ddot{\breve c}_4, \tilde{\ddot c}\\
\end{align}$$
Then the Richardson Extrapolation of the finite difference method is of order $\mathcal O(\Delta t)^4+\mathcal O(\Delta x)^4$.
